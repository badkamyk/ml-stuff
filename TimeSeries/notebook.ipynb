```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
energy = pd.read_csv('energy.csv', index_col=0)
```

```python
energy.head()
```

```python
energy.plot(y='load', subplots=True, figsize=(15, 8), fontsize=12)
plt.xlabel('timestamp', fontsize=12)
plt.ylabel('load', fontsize=12)
plt.show()
```

```python
energy['2014-07-01': '2014-07-07'].plot(y='load', subplots=True, figsize=(20, 8), fontsize=12)
plt.xlabel('timestamp', fontsize=12, labelpad=20)
plt.ylabel('load', fontsize=12)
plt.show()
```
